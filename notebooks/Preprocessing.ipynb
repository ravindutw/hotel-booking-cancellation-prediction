## Drop unnecessary columns

* Drop `reservation_status_date` and `reservation_status`

```python
df.drop(columns=['reservation_status_date', 'reservation_status'], inplace=True)
```

```python
df.shape
```

## Duplicates

```python
df[df.duplicated()].shape
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

## Missing values

```python
df.isnull().sum()
```

```python
df['agent']
```

```python
df['company']
```

### Drop missing rows in `country`

```python
df.dropna(subset=['country', 'children'], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

### Fill missing values in `agent` & `company`

```python
df.fillna(value=0, inplace=True)
```
